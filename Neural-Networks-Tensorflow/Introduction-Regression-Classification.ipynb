## Introduction to Keras

Keras is a high-level neural networks API that was originally developed as a standalone library but is now integrated into TensorFlow as `tf.keras`. It provides a user-friendly interface for building and training neural networks.

### Key Features of Keras:

1. **User-friendliness**: Keras has a simple and consistent interface, minimizing the number of user actions required for common use cases.

Let's create a simple example to demonstrate the basic structure of a Keras model:

```python
import tensorflow as tf
from tensorflow import keras

# Define a simple sequential model
def create_simple_model():
    """
    Creates and returns a simple sequential neural network model.
    
    This model consists of:
    - An input layer (flatten)
    - Two dense (fully connected) hidden layers with ReLU activation
    - An output layer with softmax activation for multi-class classification
    
    Returns:
    keras.Sequential: A compiled Keras sequential model
    """
    model = keras.Sequential([
        # Input layer: Flatten the input (assume 28x28 image)
        keras.layers.Flatten(input_shape=(28, 28)),
        
        # First hidden layer: 128 neurons with ReLU activation
        keras.layers.Dense(128, activation='relu'),
        
        # Second hidden layer: 64 neurons with ReLU activation
        keras.layers.Dense(64, activation='relu'),
        
        # Output layer: 10 neurons (e.g., for 10 classes) with softmax activation
        keras.layers.Dense(10, activation='softmax')
    ])
    
    # Compile the model
    model.compile(optimizer='adam',
                  loss='sparse_categorical_crossentropy',
                  metrics=['accuracy'])
    
    return model

# Create the model
model = create_simple_model()

# Display model summary
model.summary()
```

In this example:

1. We import TensorFlow and Keras.
2. We define a function `create_simple_model()` that creates a simple neural network using Keras' Sequential API.
3. The model has an input layer (Flatten), two hidden layers (Dense with ReLU activation), and an output layer (Dense with softmax activation).
4. We compile the model with an optimizer, loss function, and metrics.
5. Finally, we create an instance of the model and display its summary.

This structure demonstrates the basic workflow of creating a model with Keras: defining the architecture, compiling the model, and preparing it for training (although we haven't included the training step in this introduction).

As you progress through your TensorFlow journey, you'll explore more complex architectures and techniques, but this serves as a starting point to understand the basics of working with TensorFlow and Keras.

# Understanding Tensors

## Definition of Tensors

In the context of machine learning and TensorFlow, a tensor is a generalization of vectors and matrices to potentially higher dimensions. More precisely:

- A tensor is a container for data, almost always numerical data.
- It's the fundamental unit of data in TensorFlow.
- Tensors can represent various types of data, from simple scalars to complex multi-dimensional arrays.

Think of tensors as n-dimensional arrays, where n can be any non-negative integer, including 0 for scalars.

## Tensor Dimensions and Shapes

The dimensionality of a tensor is often referred to as its rank. Here's a breakdown of tensor ranks:

1. **Rank 0**: Scalar (single number)
2. **Rank 1**: Vector (1D array)
3. **Rank 2**: Matrix (2D array)
4. **Rank 3+**: Higher-dimensional array

# Tensor Ranks with Numerical Examples

## Creating and Manipulating Tensors in TensorFlow

Let's explore how to create and manipulate tensors using TensorFlow:



This code demonstrates various ways to create and manipulate tensors in TensorFlow:

1. **Creating Tensors**: We create tensors of different ranks using `tf.constant()`, from scalar to 3D tensors. We also show how to create tensors from NumPy arrays and with specific data types.

2. **Tensor Shapes**: For each tensor, we print its shape to illustrate how dimensions are represented.

3. **Special Tensors**: We create tensors filled with zeros and ones using `tf.zeros()` and `tf.ones()`.

4. **Reshaping**: We use `tf.reshape()` to change the shape of a tensor while keeping the same data.

5. **Transposing**: `tf.transpose()` is used to swap dimensions of a tensor.

6. **Slicing**: We demonstrate how to extract parts of a tensor using indexing.

7. **Basic Operations**: Addition and element-wise multiplication of tensors are shown.

8. **Matrix Multiplication**: We use `tf.matmul()` for matrix multiplication.

9. **Reduction Operations**: `tf.reduce_sum()` and `tf.reduce_max()` are used to compute the sum and maximum value of a tensor.

10. **Concatenation**: We show how to concatenate tensors along different axes using `tf.concat()`.

Understanding these operations is crucial for working with data in TensorFlow, as they form the building blocks for more complex computations in neural networks and other machine learning models.

````python

import tensorflow as tf
import numpy as np

# Creating tensors

# Scalar (rank 0)
scalar = tf.constant(42)
print("Scalar tensor:", scalar)
print("Shape:", scalar.shape)

# Vector (rank 1)
vector = tf.constant([1, 2, 3, 4, 5])
print("\nVector tensor:", vector)
print("Shape:", vector.shape)

# Matrix (rank 2)
matrix = tf.constant([[1, 2, 3], [4, 5, 6], [7, 8, 9]])
print("\nMatrix tensor:")
print(matrix)
print("Shape:", matrix.shape)

# 3D tensor
tensor_3d = tf.constant([[[1, 2], [3, 4]], [[5, 6], [7, 8]]])
print("\n3D tensor:")
print(tensor_3d)
print("Shape:", tensor_3d.shape)

# Creating tensors from NumPy arrays
np_array = np.array([[1, 2, 3], [4, 5, 6]])
tensor_from_np = tf.constant(np_array)
print("\nTensor from NumPy array:")
print(tensor_from_np)

# Creating tensors with specific data types
float_tensor = tf.constant([1.0, 2.0, 3.0], dtype=tf.float32)
int_tensor = tf.constant([1, 2, 3], dtype=tf.int32)
print("\nFloat tensor:", float_tensor)
print("Integer tensor:", int_tensor)

# Creating tensors with specific shapes
zeros_tensor = tf.zeros([3, 4])
ones_tensor = tf.ones([2, 3, 2])
print("\nZeros tensor:")
print(zeros_tensor)
print("\nOnes tensor:")
print(ones_tensor)

# Manipulating tensors

# Reshaping
reshaped_tensor = tf.reshape(matrix, [1, 9])
print("\nReshaped tensor:")
print(reshaped_tensor)

# Transposing
transposed_matrix = tf.transpose(matrix)
print("\nTransposed matrix:")
print(transposed_matrix)

# Slicing
sliced_tensor = tensor_3d[0, :, :]  # Select the first 2D slice
print("\nSliced tensor:")
print(sliced_tensor)

# Basic operations
a = tf.constant([[1, 2], [3, 4]])
b = tf.constant([[5, 6], [7, 8]])

addition = a + b
multiplication = a * b
print("\nAddition:")
print(addition)
print("\nElement-wise multiplication:")
print(multiplication)

# Matrix multiplication
matrix_mul = tf.matmul(a, b)
print("\nMatrix multiplication:")
print(matrix_mul)

# Reduce operations
sum_all = tf.reduce_sum(matrix)
max_value = tf.reduce_max(matrix)
print("\nSum of all elements:", sum_all.numpy())
print("Maximum value:", max_value.numpy())

# Concatenation
concat_horizontal = tf.concat([a, b], axis=1)
concat_vertical = tf.concat([a, b], axis=0)
print("\nHorizontal concatenation:")
print(concat_horizontal)
print("\nVertical concatenation:")
print(concat_vertical)
````

# Neural Network Basics and Key Concepts
--- 


## Understanding Input Dimensions

1. **Basic Structure**: 
   - Input data is typically structured as: n_samples * n_features
   - n_samples: Number of data points
   - n_features: Number of input variables or features for each sample

2. **Tabular Data Example**:
   - Dataset with 1000 samples, each having 5 features
   - Shape: (1000, 5)

3. **Image Data**:
   - Images have higher dimensional input
   - Typical shape: (n_samples, height, width, channels)
   - Example: 100 RGB images of size 28x28
     - Shape: (100, 28, 28, 3)

## Reshaping Data

1. **Flattening**:
   - Convert multi-dimensional data to 1D
   - Example: 
     - Original image shape: (28, 28)
     - Flattened shape: (784,)
   - Used when feeding image data into dense layers

2. **Adding Dimensions**:
   - Sometimes needed for specific layer types (e.g., convolutional layers)
   - Example:
     - Original shape: (28, 28)
     - After adding dimension: (28, 28, 1)
   - In TensorFlow: `tf.expand_dims(image, axis=-1)`

## Neural Network Structure

Neural networks typically consist of three types of layers:

1. **Input Layer**: Receives the initial data
2. **Hidden Layers**: Perform computations (can be multiple)
3. **Output Layer**: Produces the final network output

The number of neurons in the output layer depends on the task:
- Regression: Often 1 neuron (or more for multi-output regression)
- Binary Classification: 1 neuron
- Multi-class Classification: Number of neurons = Number of classes
## Key Concepts in Neural Networks

### Activation Functions

Activation functions introduce non-linearity into the network, allowing it to learn complex patterns.

#### Types and Use Cases:

1. **ReLU (Rectified Linear Unit)**:
   - f(x) = max(0, x)
   - Use case: Hidden layers in many networks, especially CNNs
   - Advantage: Helps mitigate the vanishing gradient problem

2. **Sigmoid**:
   - f(x) = 1 / (1 + e^(-x))
   - Use case: Binary classification output layer
   - Advantage: Outputs in range [0, 1], interpretable as probabilities

3. **Tanh (Hyperbolic Tangent)**:
   - f(x) = (e^x - e^(-x)) / (e^x + e^(-x))
   - Use case: Hidden layers, especially in RNNs
   - Advantage: Outputs in range [-1, 1], zero-centered

4. **Softmax**:
   - f(x_i) = e^(x_i) / Σ(e^(x_j))
   - Use case: Multi-class classification output layer
   - Advantage: Outputs sum to 1, interpretable as probabilities

5. **Leaky ReLU**:
   - f(x) = max(αx, x), where α is a small constant
   - Use case: Alternative to ReLU in hidden layers
   - Advantage: Allows small negative values, preventing "dying ReLU" problem

### Optimizers

Optimizers are algorithms or methods used to change the attributes of the neural network such as weights and learning rate to reduce the losses. 

#### Overview of Common Optimizers:

1. **Stochastic Gradient Descent (SGD)**:
   - Idea: Updates parameters in the direction of the negative gradient
   - Advantage: Simple, works well for shallow networks
   - Disadvantage: Can be slow, especially for deep networks
   - Suitable for: Simple models, large datasets
Reference Link : https://en.wikipedia.org/wiki/Stochastic_gradient_descent

2. **Adam (Adaptive Moment Estimation)**:
   - Idea: Adapts learning rates for each parameter
   - Advantage: Works well in practice, requires little tuning
   - Disadvantage: Can generalize poorly in some cases
   - Suitable for: Most deep learning applications

Reference Lik : https://builtin.com/machine-learning/adam-optimization

3. **RMSprop**:
   - Idea: Adapts learning rates using a moving average of squared gradients
   - Advantage: Good performance on non-stationary problems
   - Disadvantage: Learning rate can be sensitive
   - Suitable for: RNNs, problems with sparse gradients

Reference Link : https://deepai.org/machine-learning-glossary-and-terms/rmsprop

4. **Adagrad**:
   - Idea: Adapts learning rates based on the frequency of parameter updates
   - Advantage: Works well for sparse data
   - Disadvantage: Learning rate can become very small over time
   - Suitable for: Problems with sparse features

Reference Link : https://optimization.cbe.cornell.edu/index.php?title=AdaGrad

5. **Momentum**:
   - Idea: Adds a fraction of the previous update to the current update
   - Advantage: Helps accelerate SGD in relevant directions
   - Disadvantage: Can overshoot in some cases
   - Suitable for: Problems where SGD oscillates


The choice of optimizer can significantly impact the training speed and final performance of your neural network. It's often beneficial to experiment with different optimizers for your specific problem.

# Creating a Basic Neural Network with TensorFlow/Keras

## Understanding the Process Flow

Neural networks operate through two main processes: forward pass and backpropagation.

### Forward Pass

1. **Input Layer**: Receives the initial data.
2. **Hidden Layers**: Each neuron computes a weighted sum of its inputs, adds a bias, and applies an activation function.
3. **Output Layer**: Produces the final prediction.

Example:
```
Input → Dense Layer → Activation → Dense Layer → Activation → Output
```

### Backpropagation

1. **Compute Loss**: Compare the network's predictions to the true values.
2. **Gradient Calculation**: Compute gradients of the loss with respect to weights and biases.
3. **Update Parameters**: Adjust weights and biases to minimize the loss.

## TensorFlow/Keras Workflow

TensorFlow, with its high-level API Keras, simplifies the process of building and training neural networks. The general workflow is:

Define: You create the architecture, specifying layers and their parameters.
Compile: You set up the learning process by choosing the optimizer, loss function, and metrics.
Fit: You train the model on your data, allowing it to learn patterns and adjust its weights.
Evaluate and Predict: You assess the model's performance on unseen data and use it to make predictions.

## Key Methods in Neural Network Models

### 1. `model.build(input_shape)`

- **Purpose**: Initializes the model's weights.

### Creating Layers

The fundamental building block in Keras is the `Dense` layer. Here's how to create one:

```python
from tensorflow.keras.layers import Dense

# Creating a dense layer
dense_layer = Dense(units=64, activation='relu', input_shape=(100,))
```

Parameters of the `Dense` layer:

- `units`: Number of neurons in the layer.
- `activation`: Activation function to use (e.g., 'relu', 'sigmoid', 'tanh').
- `input_shape`: Shape of the input data (only needed for the first layer).
- `kernel_initializer`: Method for initializing weights (default is 'glorot_uniform').
- `bias_initializer`: Method for initializing biases (default is 'zeros').
- `kernel_regularizer`: Regularization method for weights (e.g., L1, L2).
- `bias_regularizer`: Regularization method for biases.


### Creating a Model

Sequential API: For linear stack of layers.

  - `batch_size`: Number of samples per gradient update.

These methods form the core of working with neural network models in frameworks like TensorFlow and Keras. Understanding and effectively using these methods is crucial for building, training, and deploying neural networks.



````python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set random seed for reproducibility
np.random.seed(0)
tf.random.set_seed(0)

"""
Generate synthetic data for our linear regression problem.
We're creating a simple linear relationship with some added noise.
"""
# X: input values, evenly spaced between 0 and 10
X = np.linspace(0, 10, 100).reshape(-1, 1)  # Reshape to column vector
# y: output values, following y = 2x + 1 with some added noise
y = 2 * X + 1 + np.random.randn(100, 1) * 0.5  # Add random noise

# Define the model
model = tf.keras.Sequential([
    # Dense layer with one neuron, suitable for simple linear regression
    tf.keras.layers.Dense(units=1, input_shape=(1,))
])

# Compile the model
model.compile(
    optimizer='adam',  # Adam optimizer, a good default choice
    loss='mean_squared_error'  # Mean Squared Error, typical for regression
)

# Train the model
history = model.fit(
    X, y,
    epochs=100,  # Number of times to iterate over the entire dataset
    batch_size=32,  # Number of samples per gradient update
    verbose=0  # Set to 0 to suppress training output
)

# Generate predictions using the trained model
y_pred = model.predict(X)

"""
Create a figure with three subplots:
1. Original data and target line
2. Neural Network predictions and target data
3. Training loss over epochs
"""
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(20, 6))
fig.suptitle('Simple Linear Regression Analysis', fontsize=16)

# Subplot 1: Original Data and Target
ax1.scatter(X, y, color='blue', alpha=0.5, label='Original Data')
ax1.plot(X, 2*X + 1, color='red', linewidth=2, label='Target Line (y = 2x + 1)')
ax1.set_title('Original Data and Target')
ax1.set_xlabel('X (Input)')
ax1.set_ylabel('y (Output)')
ax1.legend()

# Subplot 2: NN Predictions and Target Data
ax2.scatter(X, y, color='blue', alpha=0.5, label='Original Data')
ax2.plot(X, y_pred, color='green', linewidth=2, label='NN Predictions')
ax2.set_title('Neural Network Predictions')
ax2.set_xlabel('X (Input)')
ax2.set_ylabel('y (Output)')
ax2.legend()

# Subplot 3: Training Loss
ax3.plot(history.history['loss'])
ax3.set_title('Training Loss')
ax3.set_xlabel('Epoch')
ax3.set_ylabel('Mean Squared Error')

# Improve layout and display the plot
plt.tight_layout()
plt.show()

"""
After the visualization, we can print out the learned parameters
to see how close our model got to the true relationship (y = 2x + 1)
"""
weight, bias = model.layers[0].get_weights()
print(f"Learned relationship: y = {weight[0][0]:.4f}x + {bias[0]:.4f}")
print(f"True relationship: y = 2x + 1")
````

## Early Stopping

- **Purpose**: Early stopping prevents the model from continuing to train when it's no longer improving on a validation dataset.
- **How it works**: During training, the model's performance on a validation set is monitored. If the performance stops improving for a specified number of epochs (called "patience"), training is halted.
- **Benefits**: 
  1. Prevents overfitting by stopping before the model starts to memorize the training data.
  2. Saves computation time by not running unnecessary epochs.
  3. Often results in models that generalize better to unseen data.

In our case, we set a patience of 20 epochs, meaning if the validation loss doesn't improve for 20 consecutive epochs, training will stop.

## Callbacks in TensorFlow

Callbacks in TensorFlow are powerful tools that allow you to customize the behavior of your model during training. They are objects that perform actions at various stages of training (e.g., at the start or end of an epoch, before or after a batch, etc.).

Key points about callbacks:

- **Functionality**: Callbacks can modify the training process, log information, save model checkpoints, adjust learning rates, and more.
- **Types**: TensorFlow provides several built-in callbacks, including EarlyStopping, ModelCheckpoint, LearningRateScheduler, and TensorBoard.
- **Custom Callbacks**: You can also create your own custom callbacks by subclassing the `keras.callbacks.Callback` class.

In our optimized model, we use the `EarlyStopping` callback:

```python
early_stopping = tf.keras.callbacks.EarlyStopping(
    monitor='val_loss', patience=20, restore_best_weights=True
)
```

This callback:
- Monitors the validation loss (`'val_loss'`)
- Has a patience of 20 epochs
- Restores the best weights found during training

We then pass this callback to the `fit` method:

```python
history = model.fit(
    X_norm, y,
    epochs=200,
    batch_size=16,
    validation_split=0.2,
    callbacks=[early_stopping],
    verbose=0
)
```

By using callbacks like early stopping, we can create more robust training processes that automatically adapt to the data and prevent common issues like overfitting.

These changes resulted in a model that more accurately captures the true relationship in our synthetic data. The code implementation of this optimized model is provided in the following code block.

````python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set random seed for reproducibility
np.random.seed(0)
tf.random.set_seed(0)

# Generate more synthetic data
num_samples = 1000
X = np.linspace(0, 10, num_samples).reshape(-1, 1)
y = 2 * X + 1 + np.random.randn(num_samples, 1) * 0.5

'''
Feature Normalization
---------------------
Normalizing input features is a crucial step in many machine learning tasks.
It helps to ensure that all features are on a similar scale, which can lead
to faster convergence during training and can improve the performance of 
some machine learning algorithms.
'''
# Normalize the input features
X_mean, X_std = X.mean(), X.std()
X_norm = (X - X_mean) / X_std

'''
Model Definition
----------------
We Add one more layer to add more expressivity to NN
'''
# Define the model
model = tf.keras.Sequential([
    tf.keras.layers.Dense(10, activation='relu', input_shape=(1,)), 
    tf.keras.layers.Dense(1)
])

'''
Model Compilation
-----------------
We compile the model by specifying the optimizer and loss function. 
Here, we use the Adam optimizer with a lower learning rate for more
stable training, and mean squared error as our loss function.
'''
# Compile the model with a lower learning rate
optimizer = tf.keras.optimizers.Adam(learning_rate=0.001)
model.compile(optimizer=optimizer, loss='mean_squared_error')

'''
Early Stopping
--------------
We implement early stopping to prevent overfitting. This callback
will stop training when the validation loss stops improving, and
it will restore the best weights seen during training.
'''
# Implement early stopping
early_stopping = tf.keras.callbacks.EarlyStopping(
    monitor='val_loss', patience=20, restore_best_weights=True
)

'''
Model Training
--------------
We train the model on our normalized data. We use a validation split
to monitor the model's performance on unseen data during training.
The early stopping callback is used to optimize training duration.
'''
# Train the model with more epochs, smaller batch size, and early stopping
history = model.fit(
    X_norm, y,
    epochs=500,
    batch_size=16,
    validation_split=0.2,
    callbacks=[early_stopping],
    verbose=0
)

'''
Generate Predictions
--------------------
After training, we use the model to make predictions on our input data.
These predictions will be used to visualize the model's learned relationship.
'''

# Generate predictions
y_pred = model.predict(X_norm)

# Plotting
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(20, 6))
fig.suptitle('Enhanced Linear Regression Analysis', fontsize=16)

# Subplot 1: Original Data and Target
ax1.scatter(X, y, color='blue', alpha=0.5, label='Original Data')
ax1.plot(X, 2*X + 1, color='red', linewidth=2, label='True Relationship')
ax1.set_title('Original Data and True Relationship')
ax1.set_xlabel('X (Input)')
ax1.set_ylabel('y (Output)')
ax1.legend()

# Subplot 2: NN Predictions and Target Data
ax2.scatter(X, y, color='blue', alpha=0.5, label='Original Data')
ax2.plot(X, y_pred, color='green', linewidth=2, label='NN Predictions')
ax2.set_title('Neural Network Predictions')
ax2.set_xlabel('X (Input)')
ax2.set_ylabel('y (Output)')
ax2.legend()

# Subplot 3: Training Loss
ax3.plot(history.history['loss'], label='Training Loss')
ax3.plot(history.history['val_loss'], label='Validation Loss')
ax3.set_title('Training and Validation Loss')
ax3.set_xlabel('Epoch')
ax3.set_ylabel('Mean Squared Error')
ax3.legend()

plt.tight_layout()
plt.show()

# Calculate and print MSE
mse = tf.keras.losses.MeanSquaredError()(y, y_pred).numpy()
print(f"Mean Squared Error: {mse:.4f}")

# Print model summary
model.summary()
````
